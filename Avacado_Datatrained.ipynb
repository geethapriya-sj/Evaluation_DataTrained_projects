```python
df= df[['region','AveragePrice','Total Volume']]
df.head()
```

```python
df_reg = pd.get_dummies(df['region'])
df_reg
df = pd.concat([df,df_reg],axis=1)
df.head()
```

```python
#from sklearn.ensemble import RandomForestClassifer
from sklearn.metrics import precision_recall_curve
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
import warnings
warnings.filterwarnings('ignore')
```

```python
X=df.drop(columns=['AveragePrice','region'])
y=df['AveragePrice']
X.head()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
model = LinearRegression()
model.fit(X_train,y_train)
```

```python
model.score(X_test,y_test)
```

```python
from sklearn.tree import DecisionTreeRegressor
model = DecisionTreeRegressor(min_samples_split=25,criterion='friedman_mse',min_samples_leaf=5,random_state=100)
model.fit(X_train,y_train)
model.score(X_test,y_test)
```

```python
from sklearn.ensemble import RandomForestRegressor
model = DecisionTreeRegressor(min_samples_split=30,max_leaf_nodes=50)
model.fit(X_train,y_train)
model.score(X_test,y_test)
```
