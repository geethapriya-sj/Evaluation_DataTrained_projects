```python
df1 = df.copy()
for col in df.columns:
    if df1.skew().loc[col]>0.55:
        df1[col]=np.sqrt(df1[col])
```

```python
df1.skew()
```

```python
df2 = df.copy()
for col in df2.columns:
    if df2.skew().loc[col]>0.55:
        df2[col]=np.log(df2[col])
df2.skew()
```

```python
# pair plot for
sns.pairplot(df1)
```

```python
from scipy.stats import zscore
z_Score = abs(zscore(df1))
print('Original Shape:' , df1.shape)
df3 = df1.loc[(z_Score<3).all(axis=1)]
print('After removing outliers: ',df3.shape)
```

```python
df3.skew()
```

```python
X = df3.drop(columns=['W','ERA','ER','RA'])
y = df3['W']
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x=sc.fit_transform(X)
x=pd.DataFrame(x,columns=X.columns)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.1, random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
model_lin = LinearRegression()
model_lin.fit(X_train,y_train)
```

```python
predict = model_lin.predict(X_test)
```

```python
from sklearn.metrics import r2_score
r2_score(predict,y_test)
```

```python
model_lin.score(X_test,y_test)
```

```python
from sklearn.linear_model import LogisticRegression
model_log = LogisticRegression()
model_log.fit(X_train,y_train)
predict = model_log.predict(X_test)
print(model_log.score(X_test,y_test))
r2_score(predict,y_test)
```

```python
from sklearn import linear_model
reg = linear_model.Ridge()
reg.fit(X_train,y_train)
predict = reg.predict(X_test)
r2_score(y_test,predict)
```

```python
from sklearn import linear_model
reg = linear_model.Lasso()
reg.fit(X_train,y_train)
predict = reg.predict(X_test)
r2_score(y_test,predict)
```

```python
from sklearn import linear_model
reg = linear_model.LassoLars()
reg.fit(X_train,y_train)
predict = reg.predict(X_test)
r2_score(y_test,predict)
```

```python
from sklearn.model_selection import cross_val_score
cross_val_score(linear_model.Lasso(),x,y,cv=5).mean()
```

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error,r2_score,accuracy_score
final_model = LinearRegression()
final_model.fit(X_train,y_train)
predict = final_model.predict(X_test)
print(mean_absolute_error(y_test,predict))
print(mean_squared_error(y_test,predict))
print(r2_score(y_test,predict))
```

```python
import pickle
file_name='Baseball_model.pkl'
pickle.dump(final_model,open(file_name,'wb'))

final_model = pickle.load(open(file_name,'rb'))
pred = final_model.predict(X_test)
print(r2_score(y_test,pred))
```
