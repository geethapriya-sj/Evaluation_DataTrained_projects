```python
from fbprophet import Prophet
```

```python
confirmed.columns=['ds','y']
confirmed['ds'] = pd.to_datetime(confirmed['ds'])

death.columns=['ds','y']
death['ds'] = pd.to_datetime(death['ds'])

recovered.columns=['ds','y']
recovered['ds'] = pd.to_datetime(recovered['ds'])
```

```python
confirmed
```

```python
m_conformed = Prophet(interval_width=0.95)
m_conformed.fit(confirmed)
future_confirmed = m_conformed.make_future_dataframe(periods=7)
future_confirmed.tail()

m_death = Prophet(interval_width=0.95)
m_death.fit(death)
future_Death = m_death.make_future_dataframe(periods=7)
future_Death.tail()

m_recovered = Prophet(interval_width=0.95)
m_recovered.fit(recovered)
future_Recovered = m_recovered.make_future_dataframe(periods=7)
future_Recovered.tail()
```

```python
forecast_confirmed = m_conformed.predict(future_confirmed)

forecast_death = m_death.predict(future_Death)

forecast_recovery = m_recovered.predict(future_Recovered)
```
